```python
import pandas as pd
```

```python
df1 = pd.read_csv("/Users/nicholas/Downloads/farmboxtrial.csv")
```

```python
df1
```

```python
df1.iloc[0]
```

```python
l = []
for i in range(98):
    for t in df1.iloc[i][0].split():
        try:
            l.append(float(t))
        except ValueError:
            pass
```

```python
l
```

```python
humidity = []
for i in range(0,len(l),5):
    humidity.append(l[i])
```

```python
humidity
```

```python
temp = []
for i in range(1,len(l),5):
    temp.append(l[i])
```

```python
temp
```

```python
light = []
for i in range(2,len(l),5):
    light.append(l[i])
```

```python
light
```

```python
soil_moisture = []
for i in range(3,len(l),5):
    soil_moisture.append(l[i])
```

```python
soil_moisture
```

```python
time = []
for i in range(4,len(l),5):
    time.append(str(l[i])[6:10])
```

```python
time
```

```python
datetime_list = []
for i in range(4,len(l),5):
    datetime_list.append("20" + str(l[i])[0:6])
```

```python
datetime_list
```

```python
from datetime import datetime
new_datetime = []
for i in datetime_list:
    new_datetime.append(datetime.strptime(i,'%Y%m%d').strftime('%Y-%m-%d'))
```

```python
new_datetime
```

```python
df2 = {"Humidity" : humidity, "Temp" : temp, "Light" : light, "Soil Moisture" : soil_moisture, "Date" : new_datetime,
      "Time" : time}
```

```python
df2
```

```python
df2 = pd.DataFrame(data = df2)
```

```python
df2
```

```python
type(df2.index)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df2.index, "Light", data=df2)
plt.title("Light against time")
plt.xlabel("Time")
plt.ylabel("Light lux")
```

```python
plt.plot(df2.index, "Soil Moisture", data=df2)
plt.title("Soil moisture against time")
plt.xlabel("Time")
plt.ylabel("Soil Moisture")
```
